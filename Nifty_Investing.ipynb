```python
df1=df.copy()
```

```python
df1.head()
```

```python
df1['Date']=df1['Date'].astype(str)
```

```python
df1['Month'] = df1['Date'].str.split("-",expand=True)[0] + "-" + df1['Date'].str.split("-",expand=True)[1]
```

```python
intial = df1['Close'][0]
```

```python
five_per = 0
ten_per = 0
fif_per = 0
twen_per = 0
twefiv_per = 0
saving_account = 300000
portfolio_value = 0
Shares = 83
```

```python
value=[]
value.append(Shares*intial)

value1=[]
value1.append(saving_account)


for i in range(1,len(df1)):
  portfolio_value = Shares * df1['Close'][i] #it should be open
  value.append(portfolio_value)
  value1.append(saving_account)


  if df1['Month'][i] != df1['Month'][i-1]:
    saving_account = saving_account + 100000

  if df1['Close'][i] > intial:
    five_per = df1['Close'][i]*0.95
    ten_per = df1['Close'][i]*0.90
    fif_per = df1['Close'][i]*0.85
    twen_per = df1['Close'][i]*0.80
    twefiv_per = df1['Close'][i]*0.75
    intial = df1['Close'][i] # initial update
    continue

  if five_per > 0:
    if df1['Close'][i] < five_per:
      Shares = int( Shares + ((saving_account*0.20)/df1['Close'][i]))
      saving_account = (saving_account * 0.80)
      continue

  if ten_per > 0:
    if df1['Close'][i] < ten_per:
      Shares = int( Shares + ((saving_account*0.25)/df1['Close'][i]))
      saving_account = (saving_account * 0.75)
      continue

  if fif_per > 0:
    if df1['Close'][i] < fif_per:
      Shares = int( Shares + ((saving_account*0.35)/df1['Close'][i]))
      saving_account = (saving_account * 0.65)
      continue

  if twen_per > 0:
    if df1['Close'][i] < twen_per:
      Shares = int( Shares + ((saving_account*0.50)/df1['Close'][i]))
      saving_account = (saving_account * 0.50)
      continue

  if twefiv_per > 0:
    if df1['Close'][i] < twefiv_per:
      Shares = int( Shares + ((saving_account*0.70)/df1['Close'][i]))
      saving_account = (saving_account * 0.30)
      continue

```

```python
df1['portfolio_value'] = value
```

```python
df1['saving_account'] = value1
```
